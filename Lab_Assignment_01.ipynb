##### **Problem 08:**
Plot $x + y + z = 1$ for $x, y \in [-3, 3]$ with interval 0.5. Use meshgrid to generate the grid. You can use any built-in function to plot the surface.

```python

x = np.arange(-3, 3.5, 0.5)
y = np.arange(-3, 3.5, 0.5)

X, Y = np.meshgrid(x, y)

Z = 1 - X - Y


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Surface plot of x + y + z = 1')

plt.show()
```

##### **Problem 10:**
Compute 2nd order Taylor approximation of $e^{2x}$ and $e^{3x}$, and $e^{5x}$ around $a = 0$ for $x \in [-1, 1]$ with interval 0.01. Then, determine which of the following is more accurate:
> 1. Calculating the Taylor approximation separately for each $e^{2x}$, $e^{3x}$ then multiplying them together.
> 2. Calculating the Taylor approximation for $e^{5x}$.

Use mean square error (MSE) to determine the accuracy. Also, plot the Taylor approximation for 1st and 2nd method and the original function in the same plot for visual comparison.

```python
def exact_e2x(x):
    return np.exp(2 * x)

def exact_e3x(x):
    return np.exp(3 * x)

def exact_e5x(x):
    return np.exp(5 * x)

def taylor_e2x(x):
    return 1 + 2 * x + 2 * x**2

def taylor_e3x(x):
    return 1 + 3 * x + (9/2) * x**2

def taylor_e5x(x):
    return 1 + 5 * x + (25/2) * x**2

x = np.arange(-1, 1.01, 0.01)


y_e2x = exact_e2x(x)
y_e3x = exact_e3x(x)
y_e5x = exact_e5x(x)

taylor_y_e2x = taylor_e2x(x)
taylor_y_e3x = taylor_e3x(x)
taylor_y_e5x = taylor_e5x(x)

method1 = taylor_y_e2x * taylor_y_e3x

method2 = taylor_y_e5x

mse_method1 = np.mean((method1 - y_e5x)**2)
mse_method2 = np.mean((method2 - y_e5x)**2)

print(f'MSE for Method 1 (e^(2x) * e^(3x) Taylor): {mse_method1}')
print(f'MSE for Method 2 (e^(5x) Taylor): {mse_method2}')


plt.figure(figsize=(10, 6))


plt.plot(x, method1, label="Method 1: e^(2x) * e^(3x)", linestyle='--')


plt.plot(x, method2, label="Method 2: e^(5x)", linestyle=':')

plt.plot(x, y_e5x, label="Exact: e^(5x)", color='black', linewidth=2)

plt.title("Comparison of Taylor Approximations for e^(5x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
